```python
import igraph
import math
from igraph import *
g = igraph.Graph.Read_GML(r"C:\Users\Sai's PC\Downloads\Network_#healthcare_mr.gml")
```

```python
def centrality_closeness(graph):
    closeness_centralities= graph.closeness()
    return closeness_centralities
```

```python
def eigen_vec_centrality(graph):
    eigen_value=graph.evcent(directed=False, scale=True)
    return eigen_value
```

```python
close_centrality=[]
close_centrality= centrality_closeness(g)
```

```python
eigen_centrality=[]
eigen_centrality= eigen_vec_centrality(g)
```
